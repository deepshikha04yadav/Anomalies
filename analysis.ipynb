```python
# Pool scores vs aggregated
pool_total_cols = [c for c in df.columns if c.endswith(".score_breakdown.total_score")]
df["sum_pool_total"] = df[pool_total_cols].sum(axis=1, skipna=True)
df["diff_agg_vs_sum"] = df["aggregated_lp_score"] - df["sum_pool_total"]
```

```python
import matplotlib.pyplot as plt
```

```python

# Plot: histogram of differences
plt.figure(figsize=(6,4))
df["diff_agg_vs_sum"].hist(bins=50)
plt.title("Difference: Aggregated vs Sum of Pool Scores")
plt.xlabel("Score Difference")
plt.ylabel("Wallet Count")
plt.show()
```

```python
df["category_sum"] = (
    df["lp_category_breakdown.stable-stable"] +
    df["lp_category_breakdown.stable-volatile"] +
    df["lp_category_breakdown.volatile-volatile"]
)
df["diff_agg_vs_cat"] = df["aggregated_lp_score"] - df["category_sum"]
```

```python
plt.figure(figsize=(6,4))
df["diff_agg_vs_cat"].hist(bins=50, color="orange")
plt.title("Difference: Aggregated vs Category Sum")
plt.xlabel("Score Difference")
plt.ylabel("Wallet Count")
plt.show()
```

```python
deposit_cols = [c for c in df.columns if "total_deposit_all_time" in c]
df["wallet_total_deposits"] = df[deposit_cols].sum(axis=1, skipna=True)

df["deposit_decile"] = pd.qcut(df["wallet_total_deposits"].rank(method="first"), 10, labels=False)
deposit_scores = df.groupby("deposit_decile")["aggregated_lp_score"].median()
```

```python

plt.figure(figsize=(6,4))
deposit_scores.plot(kind="line", marker="o")
plt.title("Median Score by Deposit Decile")
plt.xlabel("Deposit Decile")
plt.ylabel("Median Aggregated Score")
plt.grid(True)
plt.show()
```

```python
retention_cols = [c for c in df.columns if "retained_liquidity" in c]
df["wallet_retention"] = df[retention_cols].sum(axis=1, skipna=True)

df["retention_decile"] = pd.qcut(df["wallet_retention"].rank(method="first"), 10, labels=False)
ret_scores = df.groupby("retention_decile")["aggregated_lp_score"].median()
```

```python
plt.figure(figsize=(6,4))
ret_scores.plot(kind="line", marker="o", color="green")
plt.title("Median Score by Retention Decile")
plt.xlabel("Retention Decile")
plt.ylabel("Median Aggregated Score")
plt.grid(True)
plt.show()
```

```python
# Dust & volatility correlation
dust_cols = [c for c in df.columns if "dust_deposit_count" in c or "dust_deposit_volume" in c]
vol_cols = [c for c in df.columns if "lp_volatility_stddev" in c]

df["wallet_dust"] = df[dust_cols].sum(axis=1, skipna=True)
df["wallet_volatility"] = df[vol_cols].mean(axis=1, skipna=True)
```

```python
plt.figure(figsize=(6,4))
plt.scatter(df["wallet_dust"], df["aggregated_lp_score"], alpha=0.3)
plt.title("Score vs Dust Deposits")
plt.xlabel("Dust Deposits")
plt.ylabel("Aggregated Score")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(df["wallet_volatility"], df["aggregated_lp_score"], alpha=0.3, color="red")
plt.title("Score vs Liquidity Volatility")
plt.xlabel("Volatility Std Dev")
plt.ylabel("Aggregated Score")
plt.show()
```

```python
high_dep = df["wallet_total_deposits"].quantile(0.9)
high_ret = df["wallet_retention"].quantile(0.9)
low_score = df["aggregated_lp_score"].median()
```

```python

anomalies = df[
    (df["wallet_total_deposits"] >= high_dep) &
    (df["wallet_retention"] >= high_ret) &
    (df["aggregated_lp_score"] < low_score)
].copy()
anomalies["reason"] = "High deposits + retention but low score"
```

```python
# Scatter plot to show anomalies
plt.figure(figsize=(6,4))
plt.scatter(df["wallet_total_deposits"], df["aggregated_lp_score"], alpha=0.3, label="All Wallets")
plt.scatter(anomalies["wallet_total_deposits"], anomalies["aggregated_lp_score"], color="red", label="Anomalies")
plt.title("Deposits vs Score (Anomalies Highlighted)")
plt.xlabel("Total Deposits")
plt.ylabel("Aggregated Score")
plt.legend()
plt.show()
```

```python
pool_cols = [c for c in df.columns if c.endswith("pool_id")]
out_cols = ["wallet_id","aggregated_lp_score","reason"] + pool_cols
anomalies[out_cols].to_csv("anomalies.csv", index=False)

print("Anomalies saved to anomalies.csv")
```
